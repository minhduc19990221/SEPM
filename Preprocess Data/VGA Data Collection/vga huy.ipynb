```python
import pandas as pd

data1=pd.read_excel('HanoiComputerVGA.xlsx').drop([79,105])
data2=pd.read_excel('LazadaVGA.xlsx').drop(['Location'],axis=1)
data3=pd.read_excel('PhongVuVGA.xlsx').drop(['review'],axis=1)
data4=pd.read_excel('TanThanhDanhVGA.xlsx').dropna()
data5=pd.read_excel('TikiVGA.xlsx')
data6=pd.read_excel('ViTinhNamAnVGA.xlsx')
```

```python
for index,r in data1.iterrows():
    r['price']= int(r['price'].split('đ')[0].strip().replace('.',''))

    
for index,r in data2.iterrows():
    r['price']= int(r['price'].split('₫')[0].strip().replace('.',''))

for index,r in data3.iterrows():
    r['price']= int(str(r['price']).split('đ')[0].strip().replace('.',''))

    
for index,r in data4.iterrows():
    r['price']= int(str(r['price']).split('VNĐ')[0].strip().replace('.',''))
    
    
for index,r in data5.iterrows():
    r['price']= int(str(r['price']).split('đ')[0].strip().replace('.',''))
    
for index,r in data6.iterrows():
    r['price']= int(str(r['price']).split('đ')[0].strip().replace(',',''))
    
```

```python
data = pd.concat([data1,data2,data3,data4,data5,data6]).drop_duplicates()
```

```python
data.head()
```

```python
data['imageUrl'] = data['imageUrl'].fillna('no-photo.jpeg')
a= (data['imageUrl']=='NaN').sum()
a
```

```python
count = data['product_name'].value_counts()
product = count.to_dict()
product
```

```python
values=[]
for name,count in product.items():
    datas=[]
    df = data.loc[data['product_name']==name]
    price = ""
    retailer = ""
    platform = ""


    for index, row in df.iterrows():
        price = row['price']
        retailer = row['retailer']
        platform = row['retailer']
        cat = 'VGA'
        product_url=row['storeUrl']
        image=row['imageUrl']
        
        subset ={"retailer":retailer,"platform":platform,"price":price,"image":image,"product_url":product_url}
        datas.append(subset)
    
        
    value ={"name":name,"count":count,"category":cat,
                "data":datas}
    values.append(value)   
```

```python
values
```

```python
import json
with open('fixed_vga.json', 'w') as fout:
    json.dump(values , fout)
```
