```python
count = names.value_counts()
product = count.to_dict()
product
```

```python
values=[]
for name,count in product.items():
    data=[]
    df = tv.loc[tv['product_name']==name]
    price = ""
    retailer = ""
    platform = ""


    for index, row in df.iterrows():
        price = int(row['price'])
        retailer = 'Unknown'
        platform = row['platform']
        cat = 'TVs'
        product_url=url[index]
        image=row['image_URL']
        
        subset ={"retailer":retailer,"platform":platform,"price":price,"image":image,"product_url":product_url}
        data.append(subset)
    
        
    value ={"name":name,"count":count,"category":cat,
                "data":data}
    values.append(value)   
```
